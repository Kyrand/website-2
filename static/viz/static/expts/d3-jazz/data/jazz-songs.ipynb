```python
def get_jazz_files():
    from glob import glob
    from humpy import jazzscore
    songfnames = glob('iRb_v1-0/*.jazz')
    songs = []
    for fname in songfnames:
        try:
            songs.append(jazzscore.JazzScore(fname).song)
        except:
            print 'oops - ' + str(fname)
    return songs
```

```python
def clean_chord(chord):
    chord = chord.replace(':min', 'm').replace(':maj', 'M')
    if chord[2] not in ['-', '#']:
        chord = chord[:2] + 'n' + chord[2:] 
    if chord[3] not in ['m', 'M', 'o', 'h']:
        chord = chord[:3] + 'x' + chord[3:]
    return chord
```

```python
for b in bars:
    for c in b:
        print clean_chord(c)
```
